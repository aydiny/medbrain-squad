# Synthetic Data Generation Methodology

## Overview
Training data was generated synthetically using Google Gemini Pro
as a Teacher model via the AI Studio API / Google Colab.
No real patient data was used at any stage.

---

## Phase 1: Scribe Dataset (500 rows)

**Objective:** Generate gold-standard SOAP JSON pairs from
clinical shorthand inputs.

**Model used:** Gemini 3.0 Pro (Teacher)

**Core prompt template:**
```
You are an expert clinical documentation specialist.
Convert the following raw clinical shorthand into a
structured SOAP note in strict JSON format.

Raw input: {raw_notes}

Output the following JSON schema:
{
  "subjective": "...",
  "objective": "...",
  "assessment": "...",
  "plan": "..."
}
```

**Iteration process:** Prompt was refined over 40 iterations
to ensure consistent JSON schema compliance and clinical accuracy.

---

## Phase 2: Auditor Dataset (1,000 rows)

**Objective:** Generate adversarially poisoned examples to
train the Auditor agent to detect safety failures.

**Method:** Clean Scribe outputs were programmatically
corrupted using two poisoning strategies:

### Poisoning Type 1 — PII Injection
Patient identifiers (names, workplaces, locations) were
injected back into the structured SOAP output.

### Poisoning Type 2 — Clinical Contradictions
Deliberate medical errors were introduced:
- Contraindicated medications added to treatment plan
- Diagnoses unsupported by objective findings
- Missing critical findings from assessment

**Target output format:**
```json
{
  "safety_status": "REJECTED",
  "audit_findings": ["Unmasked PII detected...",
                     "Clinical contradiction: ..."],
  "redacted_json": {...}
}
```

---

## Reproducing This Dataset

To generate your own synthetic dataset:
1. Access Google AI Studio at aistudio.google.com
2. Use Gemini Pro with the prompt templates above
3. Generate batches of 50 examples per API call
4. Apply programmatic poisoning logic to Phase 1
   outputs to create Phase 2 data

Sample inputs/outputs available in `/data/`
